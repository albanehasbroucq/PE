Lisons le fichiers 'countries.csv', et affichons les cinq premières lignes:

```python
countries = pd.read_csv('countries.csv')
countries.head()
```

Taille du data sat et noms des colonnes:

On a 214 lignes donc 214 pays avec leur données pour 8 colonnes différentes.\
Il serait alors cohérent d'indexer nos datas par le pays correspondant:

```python
countries=countries.set_index('country')
countries.tail() #affichons la fin pour changer 
```

Données statistiques de chaque colonnes :

```python
for colonne in countries.columns:
    print(countries[colonne].describe(),'\n')
```

Encore une fois les types de chaque colonne sont cohérents.
Pour chaque colonne, on a le nom de la colonne, le type des éléments qu'elle contient, le nombre d'éléments (count), le nombre d'élements différents (unique), l'élément le plus fréquent (top, pour les éléments de type objet) et sa fréquence d'apparition, et pour les colonnes dont le type est des float 64, on a le nombre de valeurs non nulles (count, les valeurs NaN ne sont pas comptées), la moyenne, l'écart-type, le minimum, les quartiles, le maximum.
